**Reviewer_Nationality:** `Reviewer_Nationality_Australia`,`Reviewer_Nationality_Ireland`,`Reviewer_Nationality_UAE`,
`Reviewer_Nationality_UK`,`Reviewer_Nationality_USA`,`Reviewer_Nationality_Other`

```python
nationalities = ['Australia', 'Ireland', 'UAE', 'UK', 'USA', 'Other']

counts_nationality = pd.DataFrame(columns=['Nationality', 'Negative_Review','Positive_Review'])

for nationality in nationalities:
    count_negative = df[(df[f'Reviewer_Nationality_{nationality}']==1) & (df['Reviewer_Score']==0)].shape[0]
    count_positive = df[(df[f'Reviewer_Nationality_{nationality}']==1) & (df['Reviewer_Score']==1)].shape[0]
    counts_nationality = counts_nationality.append({'Nationality':nationality,'Negative_Review': count_negative,'Positive_Review':count_positive},ignore_index=True)
    
counts_nationality.head()    
```

```python
#Visualize the Result
color_palette = 'Set2'


plt.figure(figsize=(10,8))
sns.barplot(data=counts_nationality, x='Nationality',y='Negative_Review', color=sns.color_palette(color_palette)[1], label='Negative Review')
sns.barplot(data=counts_nationality, x='Nationality',y='Positive_Review', color=sns.color_palette(color_palette)[0], bottom=counts_nationality['Negative_Review'],label='Positive Review')
plt.xlabel('Nationality')
plt.legend()
plt.title('Count of Positive and Negative Reviews for Each Nationality')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### 3.1.3 Text Cleaning

We can start by filtering out tokens that appear less than 10 times, which can be accomplished by setting 'min_df' to 10. Additionally, common words like 'the,' 'a,' and 'to' can be eliminated by specifying 'stop_words' as "English." To maintain textual consistency, we should exclude numbers and perform lemmatization on the words. Finally, we aim to limit the maximum number of features to 500. To achieve these requirements, a customized tokenization process can be defined.

**Customerize tockens**

```python
# import the nltk stopwords
import string
import re
import nltk

lemmatizer = nltk.WordNetLemmatizer()

nltk.download('stopwords')
from nltk.corpus import stopwords 

ENGLISH_STOP_WORDS = stopwords.words('english')

def my_tokenizer(text):
    # remove punctuation and set to lower case
    for punctuation_mark in string.punctuation:
        text = text.replace(punctuation_mark,'').lower()

    # split sentence into words

    listofwords = re.findall(r'\b\w+\b', text)
    listoflemmatized_words = []
    
    # remove stopwords and any tokens that are just empty strings
    for word in listofwords:
        if (not word in ENGLISH_STOP_WORDS) and (word != '') and (not word.isnumeric()):
            #lemmatized word
            lemmatized_word = lemmatizer.lemmatize(word)
            listoflemmatized_words.append(lemmatized_word)

    return listoflemmatized_words
```

```python
# Instantiate CountVectorizer with the custom tokenization function
neg_bagofwords = CountVectorizer(min_df=10,
                                 max_df=0.95, 
                                 tokenizer=my_tokenizer,
                                 max_features=500)

# Fit the vectorizer on your text data
neg_bagofwords.fit(negative_reviews_train['Negative Review'])

# Sanity Check of the feature names
neg_bagofwords.get_feature_names_out()[:200]
```

#### 4.3.2 Hyperparameters Tunning

As per the question's requirements, we aim to optimize at least three hyperparameters: 'max_depth,' 'min_samples_split,' and 'min_samples_leaf.' To achieve this, we will employ a GridSearch approach to identify the most suitable combination of these hyperparameters.
- **max_depth**: The first hyperparameter I choose is max_depth. This hyperparameter sets the maximum depth of the tree. A deeper tree can captrue more complex relationships in the data but may also overfit. Therefore, setting an appropriate value for max_depth helps control overfitting.
- **min_samples_split**: The min_samples_split hyperparameter in decision three specifies the minimum number of samples that must be present in a node for it to be split further into child nodes during the construction of the tree. In other words, it controls the granularity of the splits in the tree. If the number of samples in a node falls below the min_samples_split threshold, the node will not be split, and it becomes a leaf node.
- **min_samples_leaf**: The third hyperparameter is "Minimum Samples per Leaf", often denoted as min_samples_leaf, which is a crucial hyperparameter in decision tree models. It determines the minimum number of samples that must be present in a leaf node of the decision tree. This hyperparameter helps control the complexity of the tree and can be used to prevent overfitting.

```python
#find the bes
best_k = 2
best_score = 0
accsco = []

for k in range(2, 30, 2):
    estimator = [('pca', PCA(n_components=20)), ('dt', DecisionTreeClassifier(max_depth=k))]
    pipe = Pipeline(estimator)
    pipe.fit(scaled_X_train, y_train)
    sco = pipe.score(scaled_X_test, y_test)
    accsco.append(sco)
    if sco > best_score:
        best_score = sco
        best_k = k
    print(f"k = {k}, Accuracy = {sco:.2f}")
```

```python
# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(range(2, 30, 2), accsco, label="Accuracy", color='blue')
plt.scatter(best_k, best_score, color='red', marker='x', s=100, label=f"Best Point (k={best_k}, Accuracy={best_score:.2f})", zorder=5)
plt.xlabel("Max Depth (k)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Max Depth (k) for Decision Tree Classifier")
plt.legend()
plt.grid(True)
plt.show()
```
